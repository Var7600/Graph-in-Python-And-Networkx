```python
# pip install numpy networkx matplotlib
import numpy as np
import networkx as nx
import matplotlib as plt
%matplotlib inline
```

```python
# graph object
G = nx.Graph()
# list of nodes
Nodes = [1, 2, 3, 4, 5]
# list of edges
Edges = [(1,2) , (2,3), (3,4), (4,5), (1,3), (1,5)]
# call functions to add to out graph
G.add_nodes_from(Nodes)
G.add_edges_from(Edges)
# represent
nx.draw(G, with_labels=True)
```

```python
G.add_node(6)
nx.draw(G, with_labels=True)
```

```python
G.add_edge(5,6)
G.add_ede(1,6)
nx.draw(G, with_labels=True)
```

```python
# Get list of connected components
sorted(nx.connected_components(G))
```

```python
G.add_node(7)
sorted(nx.connected_components(G))
```

```python
Agraph = { 'A' : ['B' , 'C'],
          'B' : ['A' , 'C' , 'D'] ,
          'C' : ['A' , 'B' , 'D' , 'E'] ,
          'D' : ['B' , 'C' , 'E' , 'F'],
          'E' : ['C', 'D' , 'F'] ,
          'F' : ['D' , 'E'] }


graph = nx.Graph()

for node in Agraph:
    graph.add_nodes_from(node)
    for edge in Agraph[node]:
        graph.add_edge(node,edge)

liste = nx.to_dict_of_lists(graph)
print(liste)

#for visualisation

pos = {'A' : [0.00 , 0.50] , 'B' : [0.25,0.75] , 'C' : [0.25 , 0.25] , 'D' : [0.75 , 0.75],
  'E' : [0.75 , 0.25] , 'F' : [1.00 , 0.50] }

nx.draw(graph , pos , with_labels=True)
```
